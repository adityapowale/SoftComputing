```python
import cv2 as cv #for reading images
import numpy as np #for simple numpy operations
from matplotlib import pyplot as plt #for plotting images
```

```python
def getData(): # getting dataset
    dataset = [] 
    for i in range(4): # 4 images
        pic = cv.imread(str(i+1) + ".jpg" , 0) #grayscale image read
        dataset.append(pic)
    dataset = np.asarray(dataset) #converting list to numpy array
    return dataset
```

```python
dataset = getData() #getting data
```

```python
def riverData(): #getting points which correspond to  river
    data = np.empty([4 , 50]) #empty array inialization
    for i in range(4): 
        a = dataset[i][125:149,168]
        b = dataset[i][475:501,158]
        a = np.concatenate((a, b) , axis = 0)
        data[i] = a
    return data
```

```python
def nonRiverData(): #getting points which correspond to non-river
    data = np.empty([4,100]) #empty array inialization
    for i in range(4):
        a = dataset[i][314 : 324  , 301 : 311]
        a = a.flatten()
        data[i] = a
    return data
non_river_data = nonRiverData()
```

```python
river_data = riverData()
non_river_data = nonRiverData()
```

```python
plt.figure(figsize = (15 , 15)) #plotting input images
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(dataset[i] , 'gray')
plt.show()
```

```python
mean_r = np.mean(river_data , axis=1) #mean of river points
mean_r = np.reshape(mean_r, (4 , 1)) #reshaping to 4x1
mean_nr = np.mean(non_river_data , axis=1) #mean of non-river points
mean_nr = np.reshape(mean_nr , (4 , 1)) #reshaping to 4x1

cov_r = np.cov(river_data) #finding covariance of river points
cov_nr = np.cov(non_river_data) #finding covariance of non-river points
```

```python
inv_r = np.linalg.pinv(cov_r) #inverse of covarince of river points
inv_nr = np.linalg.pinv(cov_nr)  #inverse of covarince of non-river points
det_r = np.linalg.det(cov_r) #determinant of covariance of river points
det_nr = np.linalg.det(cov_nr) #determinant of covariance of non-river points
```

```python
def test(P1 , P2): #finding resultant images
    output = np.zeros([512 , 512]) #initializing a empty image
    for i in range(512): #iterating over all pixels
        for j in range(512):
            sample = dataset[: , i , j] #taking each pixel of all 4 images
            sample = np.reshape(sample , (4 , 1))
            a = sample - mean_r 
            R = np.dot(a.T , np.dot(inv_r , a))#using formula ((t - Rmean)T * (inv of river class) * (t - Rmean))
            b = sample - mean_nr
            NR = np.dot(b.T , np.dot(inv_nr , b))#using formula ((t-NRmean)T *(inv of non-river class)*(t-NRmean))
            p1 = ((-0.5) * np.exp(R[0][0])) / det_r #((-.5) * (e^R) / det(r)
            p2 = ((-0.5) * np.exp(NR[0][0])) / det_nr#((-.5) * (e^NR) / det(nr)
            if (p1* P1 >= p2*P2):
                output[i][j] = 255
            else:
                output[i][j]= 0
    return output
```

```python
out_1 = test(0.3 ,0.7)
out_2 = test(0.7 ,0.3)  #trying output with different posterior probabilities
out_3 = test(0.5 ,0.5)
```

```python
plt.figure(figsize = (15 , 15))
plt.subplot(131),plt.imshow(out_1 , 'gray')
plt.subplot(132),plt.imshow(out_2 , 'gray')  #plotting output images
plt.subplot(133),plt.imshow(out_3 , 'gray')
```
