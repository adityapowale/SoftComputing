```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housedata.csv', index_col=0)
df = df.replace({'yes': 1, 'no': 0})
df.head()
```

```python
Y = df.iloc[:, 0].values
X = df.iloc[:, 1:].values
```

```python
w = np.linalg.inv(np.matmul(X.T , X))  #(X.T * X)^-1
```

```python
y = np.matmul(X.T , Y)  #(X.T * Y)
```

```python
w = np.matmul(w , y)
```
